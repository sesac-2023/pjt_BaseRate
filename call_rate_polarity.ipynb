```python
import pandas as pd
import datetime
```

```python
#2004년~2023년 8월 콜금리
call_rate=pd.DataFrame(pd.date_range('2004-12-01','2023-08-31'), columns=["DATE"])

#실제 콜금리
call_df=pd.read_csv("./Raw_Call.csv",encoding_errors='ignore')
call_df["DATE"]=pd.to_datetime(call_df['DATE'])
call_rate=pd.merge(call_rate, call_df, how='left', on='DATE')
call_rate = call_rate.fillna(method='bfill')
```

```python
call_rate.head()
```

### 한달 전 콜금리와 비교

```python
call_rate.set_index("DATE",inplace=True, drop=False )
call_rate["MONTH_AGO"]=call_rate["DATE"].apply(lambda x: x-datetime.timedelta(days=30))

rate_ago=[]
for x in call_rate["MONTH_AGO"]:
  try:
    rate_ago.append(call_rate.loc[x,"RATE"])
  except:
    rate_ago.append(None)
call_rate["RATE_MONTH_AGO"]=rate_ago

polarity=[]
for idx, x in enumerate(call_rate["RATE_MONTH_AGO"]):
    if call_rate.iloc[idx, 1] > x:
      polarity.append(1)
    elif call_rate.iloc[idx,1] == x:
      polarity.append(0)
    else:
      polarity.append(-1)
call_rate["POLARITY"]=polarity
call_rate.dropna(inplace=True)
call_rate.drop(["MONTH_AGO"], axis=1)
```

```python
call_rate.to_csv("./call_rate_polarity.csv", index=False)
```

### 한달 후 콜금리와 비교

```python
call_rate.set_index("DATE",inplace=True, drop=False )
call_rate["MONTH_LATER"]=call_rate["DATE"].apply(lambda x: x+datetime.timedelta(days=30))

rate_ago=[]
for x in call_rate["MONTH_LATER"]:
  try:
    rate_ago.append(call_rate.loc[x,"RATE"])
  except:
    rate_ago.append(None)
call_rate["RATE_MONTH"]=rate_ago

polarity=[]
for idx, x in enumerate(call_rate["RATE_MONTH"]):
    if call_rate.iloc[idx, 1] > x:
      polarity.append(1)
    elif call_rate.iloc[idx,1] == x:
      polarity.append(0)
    else:
      polarity.append(-1)
call_rate["POLARITY"]=polarity
call_rate.dropna(inplace=True)
call_rate.drop(["MONTH_LATER"], axis=1)
```

```python
call_rate.to_csv("./call_rate_polarity(+month).csv", index=False)
```
